```python
# import pandas as pd

# df = pd.read_csv(r"C:\Users\Andrea Chiang\Downloads\python\all_classifier_batch_test\cell_coords_labeled_fixed.csv")
# counts = df['predicted_label'].value_counts()
# print(counts)

def compute_confusion_counts(N, P, R):
    """
    Compute TP, FP, FN given:
      N = total ground-truth instances
      P = precision (decimal)
      R = recall (decimal)
    """
    TP = R * N
    FN = N - TP
    FP = TP * (1 / P - 1)
    return int(round(TP)), int(round(FP)), int(round(FN))

# Define your scenarios
metrics = {
    'L ALL w/aug':    (1021, 0.983, 0.990),
    'L HEM w/aug':    (1470, 0.997, 0.987),
    'M ALL w/ aug':    (1021, 0.984, 0.99),
    'M HEM w/ aug':    (1470, 0.997, 0.989),
    'S ALL w/ aug':    (1021, 0.977, 0.977),
    'S HEM w/ aug':    (1470, 0.99, 0.982),
    'L ALL w/o aug':    (162,0.753, 0.698),
    'L HEM w/o aug':    (185,0.785, 0.768),
    'M ALL w/o aug':    (162, 0.772, 0.691),
    'M HEM w/o aug':    (185, 0.88, 0.676),
    'S ALL w/o aug':    (162, 0.759, 0.679),
    'S HEM w/o aug':    (185, 0.844, 0.703),
}

# Compute and display
for label, (N, P, R) in metrics.items():
    TP, FP, FN = compute_confusion_counts(N, P, R)
    print(f"{label:17s}: TP: {TP:4d}, FP: {FP:4d}, FN: {FN:4d}")
```

# EVALUATING MODEL 

```python
import os, glob
import pandas as pd

# ───── USER CONFIG ─────
GT_CSV       = r"C:\Users\Andrea Chiang\Downloads\python\thesis\Image Processing Models Testing\Unet\dataset\test\results\mapped_cellpose\cell_coords_labeled.csv"
PRED_GLOB   = r"C:\Users\Andrea Chiang\Downloads\python\yolo_thesis\yolov5\runs\detect\no aug\ALL_HEM_80-10-10_100Ev5mpatience20_no_aug_conf_70\labels\*.txt"
IOU_THRESH  = 0.7
OUTPUT_CSV  = r"C:\Users\Andrea Chiang\Downloads\python\yolo_thesis\ALL_HEM_80-10-10_100Ev5mpatience20_no_aug_conf_70_iou_70EVAL.csv"
# ────────────────────────

# load all GT crops
gt = pd.read_csv(GT_CSV)

# helper: convert normalized (xc,yc,w,h) → (xmin,ymin,xmax,ymax)
def to_box(xc,yc,w,h):
    return (xc - w/2, yc - h/2, xc + w/2, yc + h/2)

# IoU in normalized coords
def iou(a,b):
    xa,ya = max(a[0],b[0]), max(a[1],b[1])
    xb,yb = min(a[2],b[2]), min(a[3],b[3])
    inter = max(0, xb-xa) * max(0, yb-ya)
    area = ((a[2]-a[0])*(a[3]-a[1]) +
            (b[2]-b[0])*(b[3]-b[1]) -
            inter)
    return inter/area if area>0 else 0

results = []

for txt_path in glob.glob(PRED_GLOB):
    stem = os.path.splitext(os.path.basename(txt_path))[0]
    # stem looks like: ALL_resized_23BM66_Snap_006
    parts = stem.split("_")
    group   = "_".join(parts[0:2])    # ALL_resized
    sub     = parts[2]                # 23BM66
    slide   = parts[3] + "_" + parts[4]  # Snap_006
    source_image = f"{group}/{sub}/{slide}.jpg"

    # all GT crops on this slide
    slide_gt = gt[gt["source_image"] == source_image]
    if slide_gt.empty:
        continue

    # read all YOLO preds on this slide
    preds = []
    with open(txt_path) as f:
        for line in f:
            cls, xc,yc,w,h = line.strip().split()
            cls = int(cls)
            box = to_box(*(map(float,(xc,yc,w,h))))
            preds.append((cls, box))

    # for each GT crop match best-pred by IoU
    for _, row in slide_gt.iterrows():
        gt_box   = to_box(row.x_center, row.y_center, row.width, row.height)
        gt_label = row.predicted_label

        best_iou = 0.0
        best_cls = None
        for cls, pbox in preds:
            v = iou(gt_box, pbox)
            if v > best_iou:
                best_iou, best_cls = v, cls

        if best_iou >= IOU_THRESH:
            pred_label = "ALL" if best_cls==0 else "HEM"
        else:
            pred_label = None

        results.append({
            "crop_filename": row.crop_filename,
            "source_image":  source_image,
            "gt_label":      gt_label,
            "pred_label":    pred_label,
            "best_iou":      round(best_iou,3),
            "correct":       (pred_label == gt_label)
        })

# write out
pd.DataFrame(results).to_csv(OUTPUT_CSV, index=False)
print(f"Saved {len(results)} rows to {OUTPUT_CSV}")
```

```python
import os, glob
import pandas as pd

# ───── USER CONFIG ─────
GT_CSV      = r"C:\Users\Andrea Chiang\Downloads\python\thesis\Image Processing Models Testing\Unet\dataset\test\results\mapped_cellpose\cell_coords_labeled.csv"
PRED_GLOB   = r"C:\Users\Andrea Chiang\Downloads\python\yolo_thesis\yolov5\runs\detect\ALL_HEM_80-10-10_with_augments_100Ev5sconf70\labels\*.txt"
IOU_THRESH  = 0.7
OUTPUT_CSV  = r"C:\Users\Andrea Chiang\Downloads\python\yolo_thesis\yolov5\CSV results\ALL_HEM_80-10-10_with_augments_100Ev5s_conf_70_iou_70_EVAL.csv"
# ───────────────────────

# suffix → flips applied
SUFFIX_MAP = {
    "_vflip":  ("v",),          # vertical only
    "_hflip":  ("h",),          # horizontal only
    "_vhflip": ("v", "h"),      # vertical + horizontal
}

# ---------- helpers -----------------------------------------------------------
def split_suffix(stem):
    """return (base_stem, flips tuple) e.g. ('ALL_resized_23BM66_Snap_006', ('v',))"""
    for suf, flips in SUFFIX_MAP.items():
        if stem.endswith(suf):
            return stem[:-len(suf)], flips
    return stem, ()  # no flip

def flip_box(xc, yc, w, h, flips):
    """mirror xc / yc in YOLO space according to flips"""
    if "h" in flips:
        xc = 1.0 - xc
    if "v" in flips:
        yc = 1.0 - yc
    return xc, yc, w, h

def to_box(xc, yc, w, h):
    return (xc - w/2, yc - h/2, xc + w/2, yc + h/2)

def iou(a, b):
    xa, ya = max(a[0], b[0]), max(a[1], b[1])
    xb, yb = min(a[2], b[2]), min(a[3], b[3])
    inter  = max(0, xb - xa) * max(0, yb - ya)
    area   = ((a[2]-a[0])*(a[3]-a[1]) +
              (b[2]-b[0])*(b[3]-b[1]) - inter)
    return inter / area if area else 0
# -----------------------------------------------------------------------------

gt = pd.read_csv(GT_CSV)
results = []

for txt_path in glob.glob(PRED_GLOB):
    stem = os.path.splitext(os.path.basename(txt_path))[0]
    base_stem, flips = split_suffix(stem)          # e.g. ('...Snap_006', ('v','h'))

    # break “ALL_resized_23BM66_Snap_006” into parts
    parts = base_stem.split("_")
    if len(parts) < 5:               # safety check
        continue
    group  = "_".join(parts[0:2])    # ALL_resized
    sub    = parts[2]                # 23BM66
    slide  = parts[3] + "_" + parts[4]     # Snap_006
    source_image = f"{group}/{sub}/{slide}.jpg"

    slide_gt = gt[gt["source_image"] == source_image]
    if slide_gt.empty:
        continue

    # read YOLO detections from this file
    preds = []
    with open(txt_path) as f:
        for line in f:
            cls, xc, yc, w, h = line.strip().split()
            preds.append((int(cls), to_box(*(map(float, (xc, yc, w, h))))))

    # compare each GT crop (flipped if needed) to predictions
    for _, row in slide_gt.iterrows():
        # apply flip(s) to GT box
        xc, yc, w, h = flip_box(
            row.x_center, row.y_center, row.width, row.height, flips
        )
        gt_box = to_box(xc, yc, w, h)

        best_iou, best_cls = 0.0, None
        for cls, pbox in preds:
            v = iou(gt_box, pbox)
            if v > best_iou:
                best_iou, best_cls = v, cls

        pred_label = None
        if best_iou >= IOU_THRESH and best_cls is not None:
            pred_label = "ALL" if best_cls == 0 else "HEM"

        # create a unique crop name for augmented GT rows
        crop_name = row.crop_filename
        if flips:
            crop_name = crop_name.replace(".jpg", "_" + "".join(flips) + "flip.jpg")

        results.append({
            "crop_filename": crop_name,
            "source_image":  source_image,
            "aug_type":      "".join(flips) or "orig",
            "gt_label":      row.predicted_label,
            "pred_label":    pred_label,
            "best_iou":      round(best_iou, 3),
            "correct":       (pred_label == row.predicted_label)
        })

# save
pd.DataFrame(results).to_csv(OUTPUT_CSV, index=False)
print(f"Saved {len(results)} rows to {OUTPUT_CSV}")
```
